```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
# Loading a dataset
df = pd.read_excel("/Users/vinu/Downloads/Customer Call List.xlsx")
```

```python
#Printing first 5 rows in data frame
df.head()
```

```python
#Deleting unuseful columns from data frame
df=df.drop(columns="Not_Useful_Column") #deleting columns
df
```

```python
#Reamoving duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
# Cleaning Last_Name using strip,rstrip,lstrip methods for string column
df["Last_Name"] = df["Last_Name"].str.strip('/')
df["Last_Name"] = df["Last_Name"].str.strip('...')
df["Last_Name"] = df["Last_Name"].str.strip('_')
df["Last_Name"]
```

```python
df
```

```python
#Cleaning Phone_Number column using replace, apply and lambda methos 
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df["Phone_Number"].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('nan','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na','')
df["Phone_Number"]=df["Phone_Number"].str.replace('--','')
df
```

```python
#Dividing one Column into multiple sub columns and adding those columns into data frame
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
#Removing NaN values in data frame
df=df.fillna('')
df
```

```python
#Removing Rows which contains Phone_number values null
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)
df
```

```python
#Removing Rows which have "Do_Not_Contact" values Y
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
        df.drop(x,inplace=True)
df
```

```python
#Reseting index values
df=df.reset_index(drop=True)
df
```
